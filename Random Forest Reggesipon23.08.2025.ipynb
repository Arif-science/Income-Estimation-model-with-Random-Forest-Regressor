```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data["gender"] = le.fit_transform(data["gender"])
data["merchant_state"] = le.fit_transform(data["merchant_state"])
data['movement']=le.fit_transform(data['movement'])
data['status']=le.fit_transform(data['status'])
data['txn_description']=le.fit_transform(data['txn_description'])
data['bin_age']=le.fit_transform(data['bin_age'])
data['dow']=le.fit_transform(data['dow'])
                                        
data
```

```python
y = data['annual_salary']
x = data.drop('annual_salary', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn import metrics

def evaluate(model,X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    
    mae_test = metrics.mean_absolute_error(y_test, y_pred)
    mse_test = metrics.mean_squared_error(y_test, y_pred)
    rmse_test = np.sqrt(mse_test)
    r2_test = metrics.r2_score(y_test, y_pred)


    y_pred_train = model.predict(X_train)

    mae_train = metrics.mean_absolute_error(y_train, y_pred_train)
    mse_train = metrics.mean_squared_error(y_train, y_pred_train)
    rmse_train = np.sqrt(mse_train)
    r2_train = metrics.r2_score(y_train, y_pred_train)


    results_dict = {
        'Metric': ['MAE', 'MSE', 'RMSE', 'R2'],
        'Train': [mae_train, mse_train, rmse_train, r2_train*100],
        'Test': [mae_test, mse_test, rmse_test, r2_test*100]
    }

    results_df = pd.DataFrame(results_dict)
    
    return results_df
```

```python
reg = RandomForestRegressor(random_state=42)
reg.fit(X_train, y_train)
```

```python
from sklearn.tree import export_text

one_tree = reg.estimators_[7]

tree_rules_rf = export_text(one_tree, feature_names=list(X_train.columns))
print(tree_rules_rf)
```

```python
evaluate(reg, X_train, y_train, X_test, y_test)
```

```python
def plot_feature_importances(model, feature_names):
    
    importances = model.feature_importances_
    
    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances
    }).sort_values(by="Importance", ascending=False)
   
    
    #plt.figure(figsize=(10, 6))
    #plt.barh(importance_df['Feature'], importance_df['Importance'])
    #plt.title("Feature Importances")
    #plt.xlabel("Importance Score")
    #plt.ylabel("Feature")
    #plt.gca().invert_yaxis() 
    #plt.show()
    
    return importance_df
```

```python
feature_names = X_train.columns

feature_importance_df = plot_feature_importances(reg, feature_names)

feature_importance_df
```

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 10)]
max_depth = [int(x) for x in np.linspace(10, 100, num = 11)]

# Create the random grid
random_grid = {'n_estimators': n_estimators, # Number of trees in random forest
               'max_features': ['sqrt', 0.3, 0.5, None], # Number of features to consider at every split
               'max_depth': max_depth, # Maximum number of levels in tree
               'min_samples_split': [20,50], # Minimum number of samples required to split a node
               'min_samples_leaf': [5, 8, 10]} # Minimum number of samples required at each leaf node
```

```python
rf_random = RandomizedSearchCV(estimator = reg, param_distributions = random_grid, n_iter = 10, cv = 3, verbose=1, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train[['balance',
	'latitude',	
	'longitude',	
	'age',	
	'payment_period',	
	'month',	
	'payment_period_mean_by_bin_age',
	'day']], y_train)
```

```python
rf_random.best_params_
```

```python
optimized_model = rf_random.best_estimator_

evaluate(optimized_model, X_train[['balance',
	'latitude',	
	'longitude',	
	'age',	
	'payment_period',	
	'month',	
	'payment_period_mean_by_bin_age',
	'day']],y_train, 
  X_test[['balance',
	'latitude',	
	'longitude',	
	'age',	
	'payment_period',	
	'month',	
	'payment_period_mean_by_bin_age',
	'day']], y_test)
```

```python
from sklearn.tree import export_text

one_tree = optimized_model.estimators_[7]

tree_rules_rf = export_text(one_tree, feature_names=list(X_train[['balance',
	'latitude',	
	'longitude',	
	'age',	
	'payment_period',	
	'month',	
	'payment_period_mean_by_bin_age',
	'day']]))

print(tree_rules_rf)
```
